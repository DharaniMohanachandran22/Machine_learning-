```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Mall_Customers.csv")
df
```

```python
x=df[["Annual Income (k$)","Spending Score (1-100)"]]
```

```python
x_standardized = StandardScaler().fit_transform(x)
```

```python
dbscan=DBSCAN(eps=0.5,min_samples=5)
clusters =dbscan.fit_predict(x_standardized)
clusters
```

```python
centroids =pd.DataFrame(x_standardized).groupby(clusters).mean().values
centroids
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x["Annual Income (k$)"],x["Spending Score (1-100)"],c=clusters,cmap="viridis")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
