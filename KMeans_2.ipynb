```python
x=df.iloc[:,[3,4]].values
x
```

```python
from sklearn.cluster import KMeans
a=[]

for i in range(1,11):
  kmeans =KMeans(n_clusters=i,init='k-means++',random_state=0)
  kmeans.fit(x)
  a.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),a)
plt.xlabel("X-axis")
plt.ylabel("y-axis")
plt.show()
#to choose culters (no.culters)use lbow mwthod
```

```python
kmeansmodel =KMeans(n_clusters=5,init='k-means++',random_state=0)
y_kmeans =kmeansmodel.fit_predict(x)
```

```python
plt.scatter(x[y_kmeans ==0,0],x[y_kmeans==0,1],s=100,c='red')
plt.scatter(x[y_kmeans ==1,0],x[y_kmeans==1,1],s=100,c='green')
plt.scatter(x[y_kmeans ==2,0],x[y_kmeans==2,1],s=100,c='blue')
plt.scatter(x[y_kmeans ==3,0],x[y_kmeans==3,1],s=100,c='yellow')
plt.scatter(x[y_kmeans ==4,0],x[y_kmeans==4,1],s=100,c='black')
plt.show()
#x is a dataset.
```
