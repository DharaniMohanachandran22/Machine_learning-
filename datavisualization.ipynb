```python
import matplotlib.pyplot as plt
```

```python
x  = [1, 2, 3, 4, 5, 6, 7, 8, 9]

y1 = [1, 3, 5, 3, 1, 3, 5, 3, 1]

y2 = [2, 4, 6, 4, 2, 4, 6, 4, 2]
plt.plot(x)
```

```python
plt.plot(x,y1)
```

```python
plt.plot(x,y1,label='y1',linewidth=10,linestyle='--',color='red',marker='+')
plt.plot(x,y2,label='y2')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Multi Line Chart')
plt.legend()
plt.show()
```
